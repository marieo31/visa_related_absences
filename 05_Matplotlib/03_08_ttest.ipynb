```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import sem, ttest_ind
from random import randint
```

```python
# Read in data
wba_data = pd.read_csv("Resources/wba_data.csv")
wba_data.head()
```

```python
height_data = pd.read_csv("Resources/general_heights.csv")
height_data.head()
```

```python
# Means
meanheight_wba = np.mean(wba_data.height)
meanheight_general = np.mean(height_data.height)
```

```python
# Run the t-test
(t_stat, p) = ttest_ind(height_data.height, wba_data.height, equal_var=False)
```

```python
# Report the data
print('Mean Height of WBA: %.2f'%(meanheight_wba))
print('Mean Height (general): %.2f'%(meanheight_general))

print('p = %.2g'%(p))
if p>0.05:
    print('The difference between the two data sets is not significant')
else:
    print('The difference between the two datasets is significant')
```

```python
# Plot sample means with error bars
errors = [sem(height_data.height ), sem(wba_data.height)]
fig, ax = plt.subplots()
ax.errorbar(['General pop','WBA'],[meanheight_general, meanheight_wba] , yerr=errors, fmt='o', color='b')
fig.suptitle('Mean Height of women in general population and in the WBA population')
ax.set_xlim([-1, 2])
```
