```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import linregress
```

```python
# Read in data
df = pd.read_csv("Resources/crime_data.csv")
df.head()
```

```python
# Isolate year column
year = df.iloc[:,0]
```

```python
# Isolate the total violent crime rate, murder rate, and aggravated assault rate columns
vcrime_rate = df.iloc[:,3]
murder_rate = df.iloc[:,5]
aggr_assault_rate = df.iloc[:,11]
```

```python
# Use stats.linregress to perform a linear regression with the year and violent crime rate columns.
(a_vcrime, b_vcrime, _, _,_) = linregress(year,vcrime_rate)
```

```python
# Use stats.linregress to perform a linear regression with the year and murder rate columns.
(a_murder, b_murder, _, _,_) = linregress(year,murder_rate)
```

```python
# Finally, use stats.linregress to perform a linear regression with the year and aggravated assault rate columns.
(a_assault, b_assault, _, _,_) = linregress(year,aggr_assault_rate)
```

```python
fig = plt.figure()

plt.subplot(311)
plt.plot(year, vcrime_rate, marker='o', color='b')
plt.plot(year, a_vcrime*year+b_vcrime, '--', color='b')
plt.xticks(np.arange(1995,2012,5),[])
plt.ylabel('Violent Crime Rate')

plt.subplot(312)
plt.plot(year, murder_rate, marker='o', color='r')
plt.plot(year, a_murder*year+b_murder, '--', color='r')
plt.xticks(np.arange(1995,2012,5),[])
plt.ylabel('Murder rate')

plt.subplot(313)
plt.plot(year, aggr_assault_rate, marker='o', color='g')
plt.plot(year, a_assault*year+b_assault, '--', color='g')
plt.xticks(np.arange(1995,2012,5))
plt.ylabel('Aggr. Assault rate')


plt.show()
```
